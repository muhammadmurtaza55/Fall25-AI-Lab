```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Emotion_final.csv")

print("We have {} rows.".format(data.shape[0]))
print("We have {} columns.".format(data.shape[1]))

print("Total Null Values:", np.sum(pd.isnull(data)))


print("\nUnique Emotions:")
print(data["Emotion"].unique())

# . Filling null values using MODE
mode_value = data["Emotion"].mode()[0]
data["Emotion"] = data["Emotion"].fillna(mode_value)

#  Type conversion
data["Emotion"] = data["Emotion"].astype(str)

#  Dropping unnecessary columns (only if present)
if "id" in data.columns:
    data.drop("id", axis=1, inplace=True)


#  Check datatypes
print("\nDatatypes:")
print(data.dtypes)

# 9. Splitting into X and Y
X = data.iloc[:, 0:-1]   # text
Y = data.iloc[:, -1]     # emotion

print("\nX Shape:", X.shape)
print("Y Shape:", Y.shape)

# 10. Converting Object → Integer
Y_encoded, uniques = pd.factorize(Y)

print("\nEmotion Encoding:")
for i, emotion in enumerate(uniques):
    print(f"{emotion} → {i}")

# Replace Y with encoded values
Y = Y_encoded

print("\nFinal Y (encoded):")
print(Y[:10])
```
